```python
from pandas_datareader.data import DataReader
import numpy as np
import plotly.graph_objs as go
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import yfinance as yf
from stockstats import StockDataFrame
from keras.preprocessing.sequence import TimeseriesGenerator
```

```python
user_input = input('Input the ticker:')
df = DataReader(user_input, data_source='yahoo', start='2018-01-01', end=datetime.now())
# Create a new dataframe with only the 'Close column 
data = df.filter(['Close'])
print(data)

# Convert the dataframe to a numpy array
dataset = data.values
date = data.index
```

```python
date_number = np.linspace(1,len(date),len(date))
dataset = dataset.flatten()
mymodel = np.poly1d(np.polyfit(date_number, dataset, 3))
plt.figure(figsize=(12, 10), dpi=80)
plt.plot(date, dataset)
plt.plot(date, mymodel(date_number))
plt.show()
```

```python
from sklearn.impute import SimpleImputer

stock = StockDataFrame.retype(df)

rsi = stock['rsi']
kdj = stock['kdjk']
macd = stock['macd']
rsi = rsi[2:]
kdj = kdj[2:]
macd = macd[2:]
print(rsi,kdj,macd)

# my_imputer = SimpleImputer(strategy='constant',fill_value=0)
# reshaped_rsi = rsi.values.reshape(-1, 1)
# imputed_rsi = pd.DataFrame(my_imputer.fit_transform(reshaped_rsi))
# reshaped_kdj = kdj.values.reshape(-1, 1)
# imputed_kdj = pd.DataFrame(my_imputer.fit_transform(reshaped_kdj))
# reshaped_macd = macd.values.reshape(-1, 1)
# imputed_macd = pd.DataFrame(my_imputer.fit_transform(reshaped_macd))

# # Imputation removed date; put them back
# imputed_rsi.index = data.index
# imputed_kdj.index = data.index
# imputed_macd.index = data.index

# imputed_rsi = imputed_rsi.squeeze()
# imputed_kdj = imputed_kdj.squeeze()
# imputed_macd = imputed_macd.squeeze()
# rsi = imputed_rsi.rename('rsi')
# kdj = imputed_kdj.rename('kdj')
# macd = imputed_macd.rename('macd')
#print(rsi,kdj,macd)
```

```python
close_data_pd = pd.concat([data[2:],rsi,kdj,macd],axis=1)
#close_data_pd = close_data_pd.drop(close_data_pd.index[0:6])
close_data_pd.head()
```

```python
close_data = close_data_pd.to_numpy()
print(close_data)
```

```python
split_percent = 0.8
split = int(split_percent * len(close_data))

close_train = close_data[:split]
close_test = close_data[split:]
print(close_train)

date_train = data.index[:split]
date_test = data.index[split:]

#close_train.head()
```

```python
look_back = 5

train_generator = TimeseriesGenerator(close_train, close_train, length=look_back, batch_size=20)
test_generator = TimeseriesGenerator(close_test, close_test, length=look_back, batch_size=1)
```

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout

model = Sequential()

model.add(
    LSTM(50,
         activation='relu',
         #return_sequences=True,
         input_shape=(look_back,4))
)
model.add(Dropout(0.2))

# model.add(LSTM(128, return_sequences=True, input_shape= (date_train.shape[0], 1)))
# model.add(LSTM(64, return_sequences=False))
# model.add(Dense(25))

model.add(Dense(units=1))
model.compile(optimizer='adam', loss='mse')

num_epochs = 25
model.fit(train_generator, epochs=num_epochs, verbose=1)

prediction = model.predict(test_generator)
```

```python
# close_train = np.array(close_train)
# print(test_price)
prediction = model.predict(test_generator) + 100
prediction2 = model.predict(train_generator)
train_price = np.delete(close_train,slice(1,4),1)
test_price = np.delete(close_test,slice(1,4),1)
print(prediction)
```
