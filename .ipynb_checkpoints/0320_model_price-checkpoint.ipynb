```python
import pandas as pd
import numpy as np
import plotly.graph_objs as go
from keras.preprocessing.sequence import TimeseriesGenerator
from pandas_datareader.data import DataReader
from datetime import datetime
```

```python
user_input = input('Input the ticker:')
df = DataReader(user_input, data_source='yahoo', start='2018-01-01', end=datetime.now())
# Create a new dataframe with only the 'Close column 
data = df.filter(['Close'])
print(data)

dataset = data.values
date = data.index
```

```python
split_percent = 0.8
split = int(split_percent * len(data))

close_train = dataset[:split]
close_test = dataset[split:]

date_train = date[:split]
date_test = date[split+2:]

# close_date = df['Date']
# close_train,close_test,date_train,date_test = train_test_split(close_data,close_date,test_size=0.2)


look_back = 10

train_generator = TimeseriesGenerator(close_train, close_train, length=look_back, batch_size=20)
test_generator = TimeseriesGenerator(close_test, close_test, length=look_back, batch_size=1)
```

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout

model = Sequential()

model.add(
    LSTM(50,
         activation='relu',
         #return_sequences=True,
         input_shape=(look_back, 1))
)
model.add(Dropout(0.2))

# model.add(LSTM(128, return_sequences=True, input_shape= (date_train.shape[0], 1)))
# model.add(LSTM(64, return_sequences=False))
# model.add(Dense(25))

model.add(Dense(units=1))
model.compile(optimizer='adam', loss='mse')

num_epochs = 25
model.fit(train_generator, epochs=num_epochs, verbose=1)
#model.save('0302_model_30.h5')

prediction = model.predict(test_generator)
```

```python
close_train = np.delete(close_train,slice(1,4),1)
close_test = np.delete(close_test,slice(1,4),1)

close_train = close_train.reshape((-1))
close_test = close_test.reshape((-1))
prediction = prediction.reshape((-1))

trace1 = go.Scatter(
    x=date_train,
    y=close_train,
    mode='lines',
    name='Data'
)
trace2 = go.Scatter(
    x=date_test,
    y=prediction,
    mode='lines',
    name='Prediction'
)
trace3 = go.Scatter(
    x=date_test,
    y=close_test,
    mode='lines',
    name='Ground Truth'
)
layout = go.Layout(
    title="Stock",
    xaxis={'title': "Date"},
    yaxis={'title': "Close"}
)
fig = go.Figure(data=[trace1, trace2, trace3], layout=layout)
fig.show()
```

```python
close_test = close_test.reshape((-1))


def predict(num_prediction, model):
    prediction_list = close_test[-look_back:]

    for _ in range(num_prediction):
        x = prediction_list[-look_back:]
        x = x.reshape((1, look_back, 1))
        out = model.predict(x)[0][0]
        prediction_list = np.append(prediction_list, out)
    prediction_list = prediction_list[look_back - 1:]
    return prediction_list


def predict_dates(num_prediction):
    last_date = data.index.values[-1]
    prediction_dates = pd.date_range(last_date, periods=num_prediction + 1).tolist()
    return prediction_dates


num_prediction = 30
forecast = predict(num_prediction, model)
forecast_dates = predict_dates(num_prediction)
print(forecast)
trace4 = go.Scatter(
    x=forecast_dates,
    y=forecast,
    mode='lines',
    name='Future Prediction'
)
fig = go.Figure(data=[trace1, trace2, trace3, trace4], layout=layout)
fig.show()
```

```python
from sklearn.metrics import mean_absolute_error

score = mean_absolute_error(close_test[:-10], prediction)
print('MAE:', score)
```
