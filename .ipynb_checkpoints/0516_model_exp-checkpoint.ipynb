```python
from pandas_datareader.data import DataReader
import numpy as np
import plotly.graph_objs as go
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import yfinance as yf
from stockstats import StockDataFrame
from keras.preprocessing.sequence import TimeseriesGenerator

user_input = input('Input the ticker:')
df = DataReader(user_input, data_source='yahoo', start='2018-01-01', end=datetime.now())
# Create a new dataframe with only the 'Close column 
data = df.filter(['Close'])
print(data)

# Convert the dataframe to a numpy array
dataset = data.values
date = data.index

date_number = np.linspace(1,len(date),len(date))
dataset = dataset.flatten()
mymodel = np.poly1d(np.polyfit(date_number, dataset, 3))
plt.figure(figsize=(12, 10), dpi=80)
plt.plot(date, dataset)
plt.plot(date, mymodel(date_number))
plt.show()
```

```python
stock = StockDataFrame.retype(df)
#stock.init_all()

# print(stock)
rsi = stock['rsi_6']
kdj = stock['kdjj']
macd = stock['macdh']
boll = stock['boll']

rsi = rsi[2:]
kdj = kdj[2:]
macd = macd[2:]
boll = boll[2:]
# print(rsi)

close_data_pd = pd.concat([data[2:],boll],axis=1)
#close_data_pd = close_data_pd.drop(close_data_pd.index[0:6])

close_data_pd.head
```

```python
close_data = close_data_pd.to_numpy()

split_percent = 0.9
look_back = 10
split = int(split_percent * len(close_data))

close_train = close_data[:split]
close_test = close_data[split:]

date_train = date[:split]
date_test1 = date[split+2:]
date_test2 = date_test1.delete(slice(0,look_back))

train_generator = TimeseriesGenerator(close_train, close_train, length=look_back, batch_size=20)
test_generator = TimeseriesGenerator(close_test, close_test, length=look_back, batch_size=1)

# generator_dict = test_generator.get_config()
# generator_data = generator_dict['data']
# print(len(generator_data.split('], [')))
# print(generator_data)

from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout

model = Sequential()

model.add(
    LSTM(50,
         activation='relu',
         #return_sequences=True,
         input_shape=(look_back,close_train.shape[1]))
)
model.add(Dropout(0.2))

# model.add(LSTM(128, return_sequences=True, input_shape= (date_train.shape[0], 1)))
# model.add(LSTM(64, return_sequences=False))
# model.add(Dense(25))

model.add(Dense(units=1))
model.compile(optimizer='adam', loss='mse')

num_epochs = 25
model.fit(train_generator, epochs=num_epochs, verbose=1)
```

```python
rsi_array = rsi.to_numpy()
kdj_array = kdj.to_numpy()
macd_array = macd.to_numpy()
boll_array = boll.to_numpy()

prediction = np.array(model.predict(test_generator))
train_price = np.delete(close_train,slice(1,close_train.shape[1]),1)
test_price = np.delete(close_test,slice(1,close_train.shape[1]),1)
print('\nHead of validation results:\n',prediction[:5])
```

```python
train = train_price.reshape((-1))
test = test_price.reshape((-1))
prediction = prediction.reshape((-1))

trace1 = go.Scatter(
    x=date_train,
    y=train,
    mode='lines',
    name='Data'
)
trace2 = go.Scatter(
    x=date_test2,
    y=prediction,
    mode='lines',
    name='Prediction'
)
trace3 = go.Scatter(
    x=date_test1,
    y=test,
    mode='lines',
    name='Ground Truth'
)
layout = go.Layout(
    title="Stock",
    xaxis={'title': "Date"},
    yaxis={'title': "Close"}
)

#close_data = close_data.reshape((-1))
fig = go.Figure(data=[trace1, trace2, trace3], layout=layout)
fig.show()
```

```python
def predict(num_prediction, model):
    prediction_list = close_test[-2*look_back:]

    for i in range(num_prediction):
        x = prediction_list[-look_back:]
#         x_list = x.tolist()
#         x_series = pd.Series(x_list)
#         print(x_series)
#         x_series = np.asarray(x_series).astype('float32')
#         print(x_series)
        
        #x = x.reshape((1, look_back, 4))
        x = TimeseriesGenerator(x,x, length=look_back-1, batch_size=1)
        out = [float(model.predict(x)),0]
        #print(out)
        #prediction_list = np.append([prediction_list], out)
        prediction_list = np.vstack([prediction_list, out])
        #print(prediction_list)
        #prediction_list = prediction_list[2*look_back - 1:]
    return prediction_list


def predict_dates(num_prediction):
    last_date = data.index.values[-1]
    prediction_dates = pd.date_range(last_date, periods=num_prediction + 1).tolist()
    prediction_dates.pop(0)
    return prediction_dates



num_prediction = 10
forecast = (predict(num_prediction, model))[-10:]
forecast = np.delete(forecast,slice(1,close_train.shape[1]),1)
forecast = forecast.reshape((-1))

forecast_dates = predict_dates(num_prediction)
print(forecast)
trace4 = go.Scatter(
    x=forecast_dates,
    y=forecast,
    mode='lines',
    name='Future Prediction'
)
fig = go.Figure(data=[trace1, trace2, trace3, trace4], layout=layout)
fig.show()
```

```python
from sklearn.metrics import mean_absolute_error

score = mean_absolute_error(test_price[:-look_back], prediction)
print('MAE:', score)
```
